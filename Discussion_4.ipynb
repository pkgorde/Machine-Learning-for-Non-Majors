### Activation Function: 
A function that takes the weighted sum as input and produces the output of the perceptron. For classic perceptrons, this is often a step function that outputs either 1 (if the sum is above a certain threshold) or 0 (otherwise).

Most Common Activation functions used:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values, focusing on -2 to 2
x = np.linspace(-2, 2, 400)  # Increased the number of points for a smoother curve

# Define the functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Use np.tanh(x) directly for the tanh function

# Compute the y values for each function within the new x range
y_relu = relu(x)
y_tanh = np.tanh(x)
y_sigmoid = sigmoid(x)

# Plotting each function on a separate graph
plt.figure(figsize=(8, 6))

# Plot ReLU
plt.subplot(3, 1, 1)  # 3 rows, 1 column, 1st subplot
plt.plot(x, y_relu, label='ReLU', color='blue', linewidth=2)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.xlim([-2, 2])

# Plot Tanh
plt.subplot(3, 1, 2)  # 3 rows, 1 column, 2nd subplot
plt.plot(x, y_tanh, label='Tanh', color='red', linestyle='--', linewidth=2)
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.xlim([-2, 2])

# Plot Sigmoid
plt.subplot(3, 1, 3)  # 3 rows, 1 column, 3rd subplot
plt.plot(x, y_sigmoid, label='Sigmoid', color='green', linestyle='-.', linewidth=2)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.xlim([-2, 2])

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

### Limitations of Single Perceptron:

1. Binary Classification: The perceptron is a binary classifier, which means it predicts one of two classes. The step function in the perceptron model allows it to decide whether a given input belongs to one class or another based on the weighted sum of its inputs.
2. Learning Algorithm: The perceptron learns by adjusting its weights. The weights are updated in a way that the error between the predicted and actual class labels is minimized. For each training sample, the weights are updated using the rule

   $\omega_{new} = \omega_{old} + \mu(y - \hat y)x$

where $\mu$ is the learning rate, $y$ is the truth (target value) and $\hat y$ is the predicted label

We will discuss the rest of the variations of ANNs in upcoming classes, but lets focus on Multi-Layer Perceptron.

Unlike a single-layer perceptron, an MLP consists of multiple layers, including one input layer, one or more hidden layers, and one output layer. Each layer is made up of nodes or neurons, which are interconnected with varying weights.

The addition of hidden layers enables MLPs to learn non-linear relationships. This is a critical advancement because it allows neural networks to tackle a vast array of problems beyond linearly separable ones.

In the iris dataset, a single perceptron will not work, since the functionality of a single perceptron is binary classification only. 

Since Iris has 3 categories to be classified into, we will have to use a MLP

sgd: Stochastic Gradient Descent

```python
from sklearn.preprocessing import MinMaxScaler

X = iris.data
y = iris.target

scaler = MinMaxScaler(feature_range=(0, 1))
X_rescaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_rescaled, y, test_size=0.2, random_state=42)

# Split the dataset into training and testing sets
mlp = MLPClassifier(solver = 'sgd', random_state = 42, activation = 'logistic', learning_rate_init = 0.3, batch_size = 100, hidden_layer_sizes = (12,3), max_iter = 500)
mlp

# Train the MLP
mlp.fit(X_train, y_train)

# Make predictions
pred = mlp.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

### K-Fold cross validation

There are some inherent problems in using just one randomly split train-test split of the data and then measuring the performance on that.

K-fold cross validation is a much more robust way of calculating the performance of your model, especially if your dataset is not large, and could contain unexplainable biases.

In K-fold cross validation, the dataset is divided into 'k' equal parts. For each iteration, one part is used as the test set, and the remaining parts are used as the training set. This process is repeated 'k' times, with each part used as the test set exactly once. The final model performance is the average performance across all 'k' trials.

The main advantage of doing so is getting a reliable estimation of the model's performance on unseen data, and the ability to utilize the entire dataset for both training and testing, which is particularly beneficial for small datasets.

```python
from sklearn.model_selection import KFold

n_splits=5
kf = KFold(n_splits=n_splits)

acc = 0
mse = 0

i = 0
for train_indices, test_indices in kf.split(X):
    #print(train_indices)
    start_train, stop_train = train_indices[0], train_indices[-1]+1
    start_test, stop_test = test_indices[0], test_indices[-1]+1
    
    mlp = MLPClassifier(solver = 'sgd', random_state = 42, activation = 'logistic', learning_rate_init = 0.3, batch_size = 100, hidden_layer_sizes = (12, 3), max_iter = 500)
    mlp.fit(X[start_train:stop_train], y[start_train:stop_train])
    pred = mlp.predict(X[start_test:stop_test])
    #pred
    i+=1
    acc += accuracy_score(y[start_test:stop_test], pred)
    mse += mean_squared_error(y[start_test:stop_test], pred)
    
    print("Accuracy for batch ", i, " : ", accuracy_score(y[start_test:stop_test], pred))
    print("Mean Square Error for batch ", i, " : ", mean_squared_error(y[start_test:stop_test], pred))

print('Average Accuracy = ', acc / n_splits)
print('Average MSE = ', mse / n_splits)
```

As we can see, the answer we get here is slightly different from the normal, 1-time calculation we had done above, but this 5-fold version classification of the same dataset, with the same model is a way in which we can truly tell how well the MLP classifier does on the iris dataset.
